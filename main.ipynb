```python
# @title 👩‍🎨 Execute the 3D Photo Inpainting
%cd "image"

# Make sure all images have the proper file name and extension
from PIL import Image
import os

print("Making sure all images have the proper file name and extension ...")
for filename in os.listdir():
  new_filename = filename.lower().replace(" ", "_")

  # to lower case and replace empty spaces
  if any(new_filename.endswith(ext) for ext in ['.jpg', '.jpeg']):
    new_filename = new_filename.split(".")[0] + ".jpg"
    os.rename(filename, new_filename)
  
  # png to jpg
  if any(new_filename.endswith(ext) for ext in ['png']):
    img = Image.open(new_filename).convert("RGB")
    new_filename = new_filename.split(".")[0] + ".jpg"
    img.save(new_filename)

!rm -rf *.png

# Start inpainting
%cd ../
print("Starting 3d photo inpainting ...")
!python main.py --config argument.yml
```
